```python
import numpy as np
import matplotlib.pyplot as plt

# Load the data from the .npy files
te_full = np.load('te_final2.npy')
rho_full = np.load('rho_final2.npy')

# Ensure the two arrays have the same size
if te_full.shape != rho_full.shape:
    raise ValueError("The shapes of te_full.npy and rho_full.npy must match.")

# Function to plot the data with an optional left-bound truncation
def plot_data(left_bound=None):
    plt.figure(figsize=(8, 6))
    
    if left_bound is not None:
        # Apply the truncation based on the left bound
        mask = rho_full >= left_bound
        plt.plot(rho_full[mask], te_full[mask], label="")
    else:
        plt.plot(rho_full, te_full, label="Full Data")
    
    # Add labels and a title
    plt.xlabel('rho')
    plt.ylabel('eV')
    plt.title('T_e')
    plt.legend()
    plt.grid(False)
    plt.show()

# User input for left bound (optional)
try:
    left_bound = input("Enter a left bound for truncation (or press Enter to skip): ")
    left_bound = float(left_bound) if left_bound.strip() else None
except ValueError:
    print("Invalid input. Plotting without truncation.")
    left_bound = None

# Plot the data
plot_data(left_bound)
```
